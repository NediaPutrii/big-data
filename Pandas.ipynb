```python
def make_df(cols, ind):
    """Quickly make a DataFrame"""
    data = {c: [str(c) + str(i) for i in ind]
            for c in cols}
    return pd.DataFrame(data, ind)

# example DataFrame
make_df('ABC', range(3))
```

**Duplicate indices**

Satu perbedaan penting antara np.concatenate dan pd.concat adalah Pandas
penggabungan mempertahankan indeks, bahkan jika hasilnya akan memiliki indeks duplikat! Mempertimbangkan
contoh sederhana ini:

```python
x = make_df('AB', [0, 1])
y = make_df('AB', [2, 3])
y.index = x.index # make duplicate indices!
print(x); print(y); print(pd.concat([x, y]))
```

**Catching the repeats as an error.**

```python
try:
  pd.concat([x, y], verify_integrity=True)
except ValueError as e:
  print("ValueError:", e)
```

catch and print the error message:

```python
print(x); print(y); print(pd.concat([x, y], ignore_index=True))
```

Terkadang indeks itu sendiri tidak penting, kita sering abaikan begitu saja. Anda dapat menentukan opsi ini menggunakan flag ignore_index. Dengan
ini disetel ke True, penggabungan akan membuat indeks bilangan bulat baru untuk hasilnya Series

**Adding MultiIndex keys.**

```python
print(x); print(y); print(pd.concat([x, y], keys=['x', 'y']))
```

the result will be a hierarchically indexed series containing the
data

Secara khusus, objek GroupBy memiliki aggregate (),

```python
rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data1': range(6),
                   'data2': rng.randint(0, 10, 6)},
                  columns = ['key', 'data1', 'data2'])
df
```

**Aggregation.**

```python
df.groupby('key').aggregate(['min', np.median, max])
```

```python
df.groupby('key').aggregate({'data1': 'min',
                             'data2': 'max'})
```

**Filtering**

```python
def filter_func(x):
 return x['data2'].std() > 4
print(df); print(df.groupby('key').std());
print(df.groupby('key').filter(filter_func))
```

**Transformation.**

```python
df.groupby('key').transform(lambda x: x - x.mean())
```

**The apply() method.**

```python
def norm_by_data2(x):
  # x is a DataFrame of group values
  x['data1'] /= x['data2'].sum()
  return x
print(df); print(df.groupby('key').apply(norm_by_data2))
```

# **High-Performance Pandas: eval() and query()**

**Motivating query() and eval(): Compound Expressions**

NumPy dan Pandas mendukung operasi vektorisasi cepat; untuk
Misalnya, saat Anda menambahkan elemen dari dua array

```python
import numpy as np
rng = np.random.RandomState(42)
x = rng.rand(1000000,0)
y = rng.rand(1000000,0)
%timeit x + y
```

```python

%timeit np.fromiter((xi + yi for xi, yi in zip(x, y)),
                     dtype=x.dtype, count=len(x))
```

```python
mask = (x > 0.5) & (y < 0.5)
```

```python
tmp1 = (x > 0.5)
tmp2 = (y < 0.5)
mask = tmp1 & tmp2
```

```python
import numexpr
mask_numexpr = numexpr.evaluate('(x > 0.5) & (y < 0.5)')
np.allclose(mask, mask_numexpr)
```

**pandas.eval() for Efficient Operations**

NumPy dan Pandas mendukung operasi vektorisasi cepat; untuk
Misalnya, saat Anda menambahkan elemen dari dua larik

```python
 import pandas as pd
 import numpy as np
 nrows, ncols = 100000, 100
 rng = np.random.RandomState(42)
 df1, df2, df3, df4 = (pd.DataFrame(rng.rand(nrows, ncols))
 for i in range(4))
```

```python
 %timeit df1 + df2 + df3 + df4
```

```python
%timeit pd.eval('df1 + df2 + df3 + df4')
```

```python
np.allclose(df1 + df2 + df3 + df4,
            pd.eval('df1 + df2 + df3 + df4'))
```

**Operations supported by pd.eval()**

Pada Pandas v0.16, pd.eval () mendukung berbagai operasi. Untuk menunjukkan
ini, kita akan menggunakan bilangan bulat DataFrames berikut

```python
df1, df2, df3, df4, df5 = (pd.DataFrame(rng.randint(0, 1000, (100, 3)))
                           for i in range(5))
```

**Arithmetic operators.**

pd.aval()) support semua aritmarika

```python
 result1 = -df1 * df2 / (df3 + df4) - df5
 result2 = pd.eval('-df1 * df2 / (df3 + df4) - df5')
 np.allclose(result1, result2)
```

**Comparison operators.**

pd.aval() support semua operasi termasuk chain expression

```python
 result1 = (df1 < df2) & (df2 <= df3) & (df3 != df4)
 result2 = pd.eval('df1 < df2 <= df3 != df4')
 np.allclose(result1, result2)
```

**Bitwise operators**

pd.aval()) support the & | bitwise operator

```python
result1 = (df1 < 0.5) & (df2 < 0.5) | (df3 < df4)
result2 = pd.eval('(df1 < 0.5) & (df2 < 0.5) | (df3 < df4)')
np.allclose(result1, result2)
```

```python
result3 = pd.eval('(df1 < 0.5) and (df2 < 0.5) or (df3 < df4)')
np.allclose(result1, result3)
```

**Object attributes and indices.**

```python
result1 = df2.T[0] + df3.iloc[1]
result2 = pd.eval('df2.T[0] + df3.iloc[1]')
np.allclose(result1, result2)
```

Operasi lain, seperti pemanggilan fungsi, pernyataan bersyarat,
loop, dan konstruksi lain yang lebih terlibat, saat ini tidak diterapkan di
pd.eval (). Jika Anda ingin menjalankan jenis ekspresi yang lebih rumit ini, Anda
dapat menggunakan pustaka Numexpr itu sendiri.

**DataFrame.eval() for Column-Wise Operations**

Sama seperti Pandas memiliki fungsi pd.eval () tingkat atas, DataFrames memiliki eval ()
metode yang bekerja dengan cara serupa. Manfaat dari metode eval () adalah kolom itu
bisa disebut dengan nama.

```python
df = pd.DataFrame(rng.rand(1000, 3), columns=['A', 'B', 'C'])
df.head()
```

```python
result1 = (df['A'] + df['B']) / (df['C'] - 1)
result2 = pd.eval("(df.A + df.B) / (df.C - 1)")
np.allclose(result1, result2)
```

```python
result3 = df.eval('(A + B) / (C - 1)')
np.allclose(result1, result3)
```

**Assignment in DataFrame.eval()**

Selain opsi yang baru saja dibahas, DataFrame.eval () juga memungkinkan penetapan
ke kolom mana pun. Mari gunakan DataFrame dari sebelumnya, yang memiliki kolom 'A', 'B', dan
'C'

```python
df.head()
```

Kita dapat menggunakan df.eval () untuk membuat kolom baru 'D' dan menetapkan nilai yang dihitung
dari kolom lain:

```python
df.eval('D = (A + B) / C', inplace=True)
df.head()
```

```python
df.eval('D = (A - B) / C', inplace=True)
df.head()
```

**Local variables in DataFrame.eval()**

Metode DataFrame.eval () mendukung sintaks tambahan yang memungkinkannya bekerja
variabel Python lokal.

```python
column_mean = df.mean(1)
result1 = df['A'] + column_mean
result2 = df.eval('A + @column_mean')
np.allclose(result1, result2)
```

**DataFrame.query() Method**

DataFrame memiliki metode lain berdasarkan string yang dievaluasi, yang disebut query ()
metode.

```python
result1 = df[(df.A < 0.5) & (df.B < 0.5)]
result2 = pd.eval('df[(df.A < 0.5) & (df.B < 0.5)]')
np.allclose(result1, result2)
```

```python
result2 = df.query('A < 0.5 and B < 0.5')
np.allclose(result1, result2)
```

```python
Cmean = df['C'].mean()
result1 = df[(df.A < Cmean) & (df.B < Cmean)]
result2 = df.query('A < @Cmean and B < @Cmean')
np.allclose(result1, result2)
```

**Performance: When to Use These Functions**

Saat mempertimbangkan apakah akan menggunakan fungsi ini, ada dua pertimbangan: com‚Äê
waktu penempatan dan penggunaan memori. Penggunaan memori adalah aspek yang paling bisa diprediksi. Seperti sudah
disebutkan, setiap ekspresi gabungan yang melibatkan array NumPy atau Pandas Data
Bingkai akan menghasilkan pembuatan larik sementara secara implisit

```python
x = df[(df.A < 0.5) & (df.B < 0.5)]
```

```python
tmp1 = df.A < 0.5
tmp2 = df.B < 0.5
tmp3 = tmp1 & tmp2
x = df[tmp3]
```
